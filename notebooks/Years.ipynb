```python
def process_publishers(df):
    res = []
    for index, value in df["Publisher"].value_counts().iteritems():
        if len(res) < 6:
            res.append(index)
            res.append(value)
    return res
```

```python
def process_genres(df):
    res = []
    for index, value in df["Genre"].value_counts().iteritems():
        res.append(index)
        res.append(value)
    for genre in data["Genre"].unique():
        if genre not in res:
            res.append(genre)
            res.append(0)
    return res
```

```python
def process_platforms(df):
    return df["Platform"].mode()[0]
```

```python
def process_names(df):
    arr = []
    for name in df["Name"]:
        arr.append(len(name))
    return round(np.array(arr).mean(), 3)
```

```python
def process_sales(df):
    arr = []
    arr.append(round(df["NA_Sales"].sum(), 2))
    arr.append(round(df["EU_Sales"].sum(), 2))
    arr.append(round(df["JP_Sales"].sum(), 2))
    arr.append(round(df["Other_Sales"].sum(), 2))    
    arr.append(round(df["Global_Sales"].sum(), 2))   
    return arr
```
