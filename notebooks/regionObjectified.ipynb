```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("games.csv")
#data = data.dropna()
data.head()
```

```python
yearArrs = []
for year in range(1985, 2017):
    na, eu, jp, ot = 0,0,0,0
    for row in data[data['Year'] == year].iterrows():
        na += row[1]["NA_Sales"]
        eu += row[1]["EU_Sales"]
        jp += row[1]["JP_Sales"]
        ot += row[1]["Other_Sales"]
    na = round(na, 2)
    eu = round(eu, 2)
    jp = round(jp, 2)
    ot = round(ot, 2)
    na_lst = [str(year) + ".na", na]
    eu_lst = [str(year) + ".eu", eu]
    jp_lst = [str(year) + ".jp", jp]
    ot_lst = [str(year) + ".ot", ot]
    yearArrs.append([str(year)])
    yearArrs.append(na_lst)
    yearArrs.append(eu_lst)
    yearArrs.append(jp_lst)
    yearArrs.append(ot_lst)
yearArrs
```

```python
df = pd.DataFrame(yearArrs, columns = ["id", "value"])
df
```

```python
df.to_csv("region_sales_obj.csv", index=False)
```
