```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("games.csv")
data.head()
```

```python
scores = data.dropna()
scores
```

```python
scores.shape
```

```python
scores["Score_Diff"].unique()
```

```python
def get_features():
    score_array = []
    for score in scores["Score_Diff"].unique():
        user_scores, critic_scores, global_sales, genres = [], [], [], []
        
        df = data[data["Score_Diff"] == score]
        for index, row in df.iterrows():
            user_scores.append(row["User_Score"])
            critic_scores.append(row["Critic_Score"])
            global_sales.append(row["Global_Sales"])
            genres.append(row["Genre"])
        
        score_array.append([score, df.shape[0], round(np.array(critic_scores).mean(), 2), round(np.array(user_scores).mean(), 2), np.array(global_sales).sum()])
    return score_array
```

```python
df = pd.DataFrame(get_features(), columns=["Score_Diff", "Count", "Critic_Avg", "User_Avg", "Global_Sales"]).sort_values(by="Score_Diff")
```

```python
df.to_csv("scores.csv", index=False)
```
